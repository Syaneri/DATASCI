### Load the page

```python
import requests

URL="https://newsapi.org/v2/everything?q=tesla&from=2021-02-19&sortBy=publishedAt&apiKey=47934c4795874c1f89733b3ed6549943"
```

### Query

```python
query = { "sortBy": "top", "from": "2021-03-11", "to": "2021-03-12", "apiKey": "47934c4795874c1f89733b3ed6549943" }
page = requests.get(URL, params = query)
```

### Turn Data into JSON format

```python
article_list = page.json()
article_list
```

### Get Articles

```python
articles = article_list["articles"]
articles
```

### Save into JSON

```python
import json

filename = "news_articles.json"
news_file = open(filename, 'w')
json.dump(articles, news_file, indent = 2)
news_file.close()
```
